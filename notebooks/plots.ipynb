```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df_generalization = pd.read_csv("../out/generalization/4uc51x21.csv").groupby(by=["scale", "simulation_idx"]).mean().drop(columns=["step_idx"]).reset_index()
df_generalization.head()
```

```python
sns.catplot(data=df_generalization, x="scale", y="mse", kind="point", errorbar=("sd", 1))
plt.ylabel("MSE for 100 experiments")
plt.xlabel("Scale (km)")
plt.title("Generalization to unseen scales")
plt.show()
```
